```python
# Work with Data
import pandas as pd
import numpy as np

path="./clients.csv"
df=pd.read_csv(path)
df.head() 
```

```python
# Get general data for each column
print(df.isnull().sum())
df.describe()
```

```python
# We need to fill na objects in order to get correlation better down the line
df.fillna(0, inplace=True)
print(df.isnull().sum())
# Get the relationship between variables
df.corr().round(2)
```

```python
import matplotlib.pyplot as plt 
plt.boxplot(df, labels=df.columns, showmeans=True)
plt.show()
```

```python
import seaborn as sns
corrs=df.corr()
sns.set(rc={'figure.figsize':(15,10)})
sns.heatmap(corrs, vmin=-1, vmax= 1, cmap="BuGn", annot= True)
```

```python
# Start by selecting variables in importance 
# Selected X12-x17 are very related variables we can take any of them
cols=['X12','X13','X14','X15','X16','X17']
ndf = df[cols]
print(ndf.corr())

Varience = np.cov([df.X12, df.X13,df.X14, df.X15, df.X16, df.X17])
Varience
```

```python
ndf = df[cols]
ndf.var().sum()
print(ndf.var())
t= ndf.var().sum()
var16 =Varience[0][0]
var17 = Varience[1][1]
print('Total Varience \t\t\t:', t.round(2) )
print('Varience prop. of X16 \t:', ((var16/t)*100).round(2),'%' )
print('Varience prop. of X17 \t:', ((var17/t)*100).round(2),'%' )
```

```python
import seaborn as sns
p = sns.regplot(data = df,  
                x = df.X16, 
                y = df.X17
               ) 
p
```

```python
import numpy as np
from sklearn.decomposition import PCA
pcs = PCA(n_components=6)

cols2=['PC1', 'PC2','PC3','PC4','PC5','PC6']
pcs.fit(df[cols])

pcsSummary = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                          'Proportion of variance': pcs.explained_variance_ratio_,
                          'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)#column 
                          }
                          )
pcsSummary = pcsSummary.transpose()
pcsSummary.round(2)

pcsSummary.columns = cols2
pcsSummary = pcsSummary.round(2)
pcsSummary
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
PC_components = np.arange(pcs.n_components_) + 1
#PC_components

_ = sns.set(style = 'whitegrid', 
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components, 
                y = pcs.explained_variance_ratio_, 
                color = 'b'
                )

_ = sns.lineplot(x = PC_components-1, 
                 y = np.cumsum(pcs.explained_variance_ratio_), 
                 color = 'black', 
                 linestyle = '-', 
                 linewidth = 2, 
                 marker = 'o', 
                 markersize = 8
                 )

plt.title('Scree Plot')
plt.xlabel('N-th Principal Component')
plt.ylabel('Variance Explained')
plt.ylim(0, 1)
plt.show()
```

```python
pcs.components_.round(4)
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(), 
                                columns=cols2,
                                index=cols
                                )
pcsComponents_df

scores = pd.DataFrame(pcs.transform(df[cols]),
                      columns=['x', 'y','z','w','u','v']
                      )
scores.head()
```

```python

pcsSummary_df.columns= ['PC{}'.format(i) for i in range(1, len(pcsSummary_df.columns)+1)]
pcsCompsDf = pd.DataFrame(pcs.components_.transpose(),
                         columns=pcsSummary_df.columns,
                         index= corrs.columns
)
pcsCompsDf.iloc[:,:5]
print(pcsCompsDf.PC1.abs().nlargest(3))
print(pcsCompsDf.PC2.abs().nlargest(3))
print(pcsCompsDf.PC3.abs().nlargest(3))
```

```python
# Use scaleing to get a better result.
from sklearn import preprocessing
pcs=PCA()
pcs.fit(preprocessing.scale((df)))
pcsSummary_df = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                             'Proportion of variance': pcs.explained_variance_ratio_,
                             'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)
                             })
pcsSummary_df = pcsSummary_df.transpose()

PC_components = np.arange(pcs.n_components_) + 1
#PC_components

_ = sns.set(style = 'whitegrid', 
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components, 
                y = pcs.explained_variance_ratio_, 
                color = 'b'
                )

_ = sns.lineplot(x = PC_components-1, 
                 y = np.cumsum(pcs.explained_variance_ratio_), 
                 color = 'black', 
                 linestyle = '-', 
                 linewidth = 2, 
                 marker = 'o', 
                 markersize = 8
                 )

plt.title('Scree Plot')
plt.xlabel('N-th Principal Component')
plt.ylabel('Variance Explained')
plt.ylim(0, 1)
plt.show()
```
